```python
import numpy as np
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from faker import Faker
faker = Faker()
```

```python
userRating = pd.read_csv('InputFoodData.csv')
```

```python
userRating
```

```python
userRating['category'].value_counts()
```

```python
users = userRating['userID'].unique()
```

```python
type(users)
```

```python
users
```

```python
name = []
for i in users:
    x = faker.name()
    name.append(x)
```

```python
name
```

```python
username = pd.DataFrame(list(zip(users,name)),columns=['userID','userName'])
```

```python
username
```

```python
userData = pd.merge(userRating,username,on='userID',how='inner')
```

```python
userData.head()
```

```python
userData.to_csv('InputName.csv')
```
