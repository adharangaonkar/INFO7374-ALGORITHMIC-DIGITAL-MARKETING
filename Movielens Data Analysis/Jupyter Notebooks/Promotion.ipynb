```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_stats = pd.read_csv("allstats_full.csv")
```

```python
all_stats
```

```python
all_stats['year']= all_stats.title.str.extract("\((\d{4})\)",expand =True) 
```

```python
all_stats['year'] = pd.to_numeric(all_stats['year'])
```

```python
all_stats
```

```python
all_stats.dropna()
```

## Adding constraints to get only the relevant data

```python
all_stats_recent = all_stats[all_stats.year >2014]
```

```python
all_stats_recent
```

```python
all_stats_recent = all_stats_recent[all_stats.ratings_count> 6000]
```

```python
all_stats_recent
```

```python
all_stats_recent = all_stats_recent[all_stats_recent.ratings_mean > 3.9]
```

```python
all_stats_recent
```

```python
len(all_stats_recent)
```

```python
plt.figure(figsize=(14,10))
plt.scatter(all_stats_recent.ratings_mean, all_stats_recent.ratings_count)
```

```python
new_ids1 =  pd.DataFrame(set(all_stats_recent.movieId), columns=['movieId'])
new_ids1
```

# Combining new movies with all ratings

```python
ratings = pd.read_csv("rating_date.csv")
```

```python
df = ratings[ratings.set_index(['movieId']).index.isin(new_ids1.set_index(['movieId']).index)]
```

```python
df
```

```python
len(set(df.movieId))
```

```python
df = df[["userId", "movieId", "rating", "timestamp"]]
df
```

```python
df.to_csv("test2.csv", index=False)
```
