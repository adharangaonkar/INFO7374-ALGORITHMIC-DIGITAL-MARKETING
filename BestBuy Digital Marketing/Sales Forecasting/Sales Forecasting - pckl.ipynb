# Import Product Data

```python
df['Daily Sales'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
df_daily_sales = df.groupby('Date')['Daily Sales'].sum().reset_index()
```

```python
# df_daily_sales2 = df_daily_sales[:-40]
```

```python
# df_daily_sales2
```

```python
df_daily_sales = df_daily_sales[:-10]
```

```python
df_daily_sales
```

```python
df_daily_sales = df_daily_sales.set_index('Date')
```

```python
df_daily_sales
```

```python
# df2 = df_daily_sales[df_daily_sales['Date']<]
```

```python
daily_df = df_daily_sales.resample('D').mean()
```

```python
daily_df
```

```python
# daily_df.to_csv("daily_df.csv")
```

```python
d_df = daily_df.reset_index().dropna()
```

```python
d_df.columns = ['ds', 'y']
fig = plt.figure(facecolor='w', figsize=(20, 6))
plt.plot(d_df.ds, d_df.y)
```

```python
m = Prophet()
m.fit(d_df)
future = m.make_future_dataframe(periods=30)
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
forecast.to_csv('SaleForecast.csv')
```

```python
import pickle
with open('forecast_model.pckl', 'wb') as fout:
    pickle.dump(m, fout)
```

```python
from datetime import datetime, timedelta
fig1 = m.plot(forecast)
#datenow = datetime.now()
datenow = datetime(2019,11,28)
dateend = datenow + timedelta(days=30)
datestart = dateend - timedelta(days=700)
plt.xlim([datestart, dateend])
plt.title("Iron/steel forecast", fontsize=20)
plt.xlabel("Day", fontsize=20)
plt.ylabel("Iron/steel price", fontsize=20)
plt.axvline(datenow, color="k", linestyle=":")
plt.show()
```
