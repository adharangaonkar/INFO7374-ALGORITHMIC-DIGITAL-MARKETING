```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
mobile_names = []
mobile_prices = []
mobile_img = []
```

```python
for i in range(1,32):
    print("i = ", i)

    page = requests.get("https://www.newegg.com/p/pl?N=100161551%204814%2050001759%2050107909%2050001077%2050001514%2050079589&cm_sp=Cat_Cell-Phones_1-_-VisNav-_-Unlocked-Cell-Phones&page="+str(i))


    soup = BeautifulSoup(page.content, 'html.parser')
    items = soup.find_all(class_="item-cell")
    

    for item in items:
        try:
            temp_name = item.find(class_='item-title').get_text()
            mobile_names.append(temp_name)
        except:
            mobile_names.append('')

    for item in items:
        try:
            temp_price = item.find(class_='price-current').get_text()
            mobile_prices.append(temp_price)
        except:
            mobile_prices.append('')

    for item in items:
        try:
            temp_src = item.find('img')
            mobile_img.append(temp_src['src'])
        except:
            mobile_prices.append('')

```

```python
mobile_df = pd.DataFrame(
    {'names': mobile_names, 'prices': mobile_prices, 'img': mobile_img})

print(mobile_df)

print("Export Done")
```

```python
mobile_df['type'] = "Unlocked Mobiles"
```
